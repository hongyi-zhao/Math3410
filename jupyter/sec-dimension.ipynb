<div class="mathbook-content"><h6 class="heading hide-type"><span class="type">Section</span> <span class="codenumber">6</span> <span class="title">Basis and dimension</span></h6><a href="sec-dimension.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-80">Next, we begin with an important result, sometimes known as the “Fundamental Theorem”:</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-steinitz"><h6 class="heading"><span class="type">Theorem</span> <span class="codenumber">6.1</span>. <span class="title">Fundamental Theorem (Steinitz Exchange Lemma).</span></h6><p id="p-81">Suppose $V = \spn\{\vec{v}_1,\ldots, \vec{v}_n\}\text{.}$ If $\{\vec{w}_1,\ldots, \vec{w}_m\}$ is a linearly independent set of vectors in $V\text{,}$ then $m\leq n\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-82">If a set of vectors spans a vector space $V\text{,}$ and it is not independent, we observed that it is possible to remove a vector from the set and still span $V\text{.}$ This suggests that spanning sets that are also linearly independent are of particular importance, and indeed, they are important enough to have a name.</p></div>

<div class="mathbook-content"><article class="definition-like" id="def-basis"><h6 class="heading"><span class="type">Definition</span> <span class="codenumber">6.2</span>.</h6><p id="p-83">Let $V$ be a vector space. A set $\mathcal{B}=\{\vec{e}_1,\ldots, \vec{e}_n\}$ is called a <dfn class="terminology">basis</dfn> of $V$ if $\mathcal{B}$ is linearly independent, and $\operatorname{span}\mathcal{B} = V\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-84">The importance of a basis is that vector vector $\vec{v}\in V$ can be written in terms of the basis, and this expression as a linear combination of basis vectors is <em class="emphasis">unique</em>. Another important fact is that every basis has the same number of elements.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="thm-invariance"><h6 class="heading"><span class="type">Theorem</span> <span class="codenumber">6.3</span>. <span class="title">Invariance Theorem.</span></h6><p id="p-85">If $\{\vec{e}_1,\ldots, \vec{e}_n\}$ and $\{\vec{f}_1,\ldots, \vec{f}_m\}$ are both bases of a vector space $V\text{,}$ then $m=n\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-86">Suppose $V=\spn\{\vec{v}_1,\ldots,\vec{v}_n\}\text{.}$ If this set is not linearly independent, <a href="sec-span.ipynb#theorem-surplus-span" class="internal" title="Theorem 4.5">Theorem 4.5</a> tells us that we can remove a vector from the set, and still span $V\text{.}$ We can repeat this procedure until we have a linearly independent set of vectors, which will then be a basis. These results let us make a definition.</p></div>

<div class="mathbook-content"><article class="definition-like" id="def-dimension"><h6 class="heading"><span class="type">Definition</span> <span class="codenumber">6.4</span>.</h6><p id="p-87">Let $V$ be a vector space. If $V$ can be spanned by a finite number of vectors, then we call $V$ a <dfn class="terminology">finite-dimensional</dfn> vector space. If $V$ is finite-dimensional (and non-trivial), and $\{\vec{e}_1,\ldots, \vec{e}_n\}$ is a basis of $V\text{,}$ we say that $V$ has <dfn class="terminology">dimension</dfn> $n\text{,}$ and write</p><div class="displaymath">
\begin{equation*}
\dim V = n\text{.}
\end{equation*}
</div><p>If $V$ cannot be spanned by finitely many vectors, we say that $V$ is <dfn class="terminology">infinite-dimensional</dfn>.</p></article></div>

Since there are two vectors in this basis, $\dim U_1 = 2\text{.}$</p></li><li id="li-9"><p id="p-95">If $p(1)=0\text{,}$ then $p(x)=(x-1)q(x)$ for some polynomial $q\text{.}$ Since $U_2$ is a subspace of $P_2\text{,}$ the degree of $q$ is at most 2. Therefore, $q(x)=ax+b$ for some $a,b\in\R\text{,}$ and</p><div class="displaymath">
\begin{equation*}
p(x) = (x-1)(ax+b) = a(x^2-x)+b(x-1)\text{.}
\end{equation*}
</div><p>Since $p$ was arbitrary, this shows that $U_2 = \spn\{x^2-x,x-1\}\text{.}$</p><p id="p-96">The set $\{x^2-x,x-1\}$ is also independent, since neither vector is a scalar multiple of the other. Therefore, this set is a basis, and $\dim U_2=2\text{.}$</p></li><li id="li-10"><p id="p-97">If $p(x)=p(-x)\text{,}$ then $p(x)$ is an even polynomial, and therefore $p(x)=a+bx^2$ for $a,b\in\R\text{.}$ (If you didn't know this it's easily verified: if</p><div class="displaymath">
\begin{equation*}
a+bx+cx^2 = a+b(-x)+c(-x)^2\text{,}
\end{equation*}
</div><p>we can immediately cancel $a$ from each side, and since $(-x)^2=x^2\text{,}$ we can cancel $cx^2$ as well. This leaves $bx=-bx\text{,}$ or $2bx=0\text{,}$ which implies that $b=0\text{.}$)</p><p id="p-98">It follows that the set $\{1,x^2\}$ spans $U_3\text{,}$ and since this is a subset of the standard basis $\{1,x,x^2\}$ of $P_2\text{,}$ it must be independent, and is therefore a basis of $U_3\text{,}$ letting us conclude that $\dim U_3=2\text{.}$</p></li></ol></div></div></div></article></div>

<div class="mathbook-content"><p id="p-99">We've noted a few times now that if $\vec{w}\in\spn\{\vec{v}_1,\ldots, \vec{v}_n\}\text{,}$ then</p><div class="displaymath">
\begin{equation*}
\spn\{\vec{w},\vec{v}_1,\ldots, \vec{v}_n\}=\spn\{\vec{v}_1,\ldots, \vec{v}_n\}
\end{equation*}
</div><p>If $\vec{w}$ is not in the span, we can make another useful observation:</p></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-independent"><h6 class="heading"><span class="type">Lemma</span> <span class="codenumber">6.10</span>. <span class="title">Independent Lemma.</span></h6><p id="p-100">Suppose $\{\vec{v}_1,\ldots, \vec{v}_n\}$ is a linearly independent set of vectors in a vector space $V\text{.}$ If $\vec{u}\in V$ but $\vec{u}\notin \spn\{\vec{v}_1,\ldots, \vec{v}_n\}\text{,}$ then $\{\vec{u},\vec{v}_1,\ldots, \vec{v}_n\}$ is independent.</p></article><article class="proof" id="proof-1"><h6 class="heading"><span class="type">Proof.</span></h6><p id="p-101">Suppose $S=\{\vec{v}_1,\ldots, \vec{v}_n\}$ is independent, and that $\vec{u}\notin\spn S\text{.}$ Suppose we have</p><div class="displaymath">
\begin{equation*}
a\vec{u}+c_1\vec{v}_1+c_2\vec{v}_2+\cdots +c_n\vec{b}_n=\vec{0}
\end{equation*}
</div><p>for scalars $a,c_1,\ldots, c_n\text{.}$ We must have $a=0\text{;}$ otherwise, we can multiply by $\frac1a$ and rearrange to obtain</p><div class="displaymath">
\begin{equation*}
\vec{u} = -\frac{c_1}{a}\vec{v}_1-\cdots -\frac{c_n}{a}\vec{v}_n\text{,}
\end{equation*}
</div><p>but this would mean that $\vec{u}\in \spn S\text{,}$ contradicting our assumption.</p><p id="p-102">With $a=0$ we're left with</p><div class="displaymath">
\begin{equation*}
c_1\vec{v}_1+c_2\vec{v}_2+\cdots +c_n\vec{b}_n=\vec{0}\text{,}
\end{equation*}
</div><p>and since we assumed that the set $S$ is independent, we must have $c_1=c_2=\cdots=c_n=0\text{.}$ Since we already showed $a=0\text{,}$ this shows that $\{\vec{u},\vec{v}_1,\ldots, \vec{v}_n\}$ is independent.</p></article></div>

<div class="mathbook-content"><p id="p-103">This is, in fact, an “if and only if” result. If $\vec{u}\in\spn\{\vec{v}_1,\ldots, \vec{v}_n\}\text{,}$ then $\{\vec{u},\vec{v}_1,\ldots, \vec{v}_n\}$ is not independent. Above, we argued that if $V$ is finite dimensional, then any spanning set for $V$ can be reduced to a basis. It probably won't surprise you that the following is also true.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="lem-enlarge-independent"><h6 class="heading"><span class="type">Lemma</span> <span class="codenumber">6.11</span>.</h6><p id="p-104">Let $V$ be a finite-dimensional vector space, and let $U$ be any subspace of $V\text{.}$ Then any independent set of vectors $\{\vec{u}_1,\ldots, \vec{u}_k\}$ in $U$ can be enlarged to a basis of $U\text{.}$</p></article><article class="proof" id="proof-2"><h6 class="heading"><span class="type">Proof.</span></h6><p id="p-105">This follows from <a href="sec-dimension.ipynb#lemma-independent" class="internal" title="Lemma 6.10: Independent Lemma">Lemma 6.10</a>. If our independent set of vectors spans $U\text{,}$ then it's a basis and we're done. If not, we can find some vector not in the span, and add it to our set to obtain a larger set that is still independent. We can continue adding vectors in this fashion until we obtain a spanning set.</p><p id="p-106">Note that this process <em class="emphasis">must</em> terminate: $V$ is finite-dimensional, so there is a finite spanning set for $V\text{,}$ and therefore for $U\text{.}$ By the Steinitz Exchange lemma, our independent set cannot get larger than this spanning set.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="thm-basis-exist"><h6 class="heading"><span class="type">Theorem</span> <span class="codenumber">6.12</span>.</h6><p id="p-107">Any finite-dimensional vector space $V$ has a basis. Moreover:</p><ol class="decimal"><li id="li-11"><p id="p-108">If $V$ can be spanned by $m$ vectors, then $\dim V\leq m\text{.}$</p></li><li id="li-12"><p id="p-109">Given an independent set $I$ in $V\text{,}$ and a basis $\mathcal{B}$ of $V\text{,}$ we can enlarge $I$ to a basis of $V$ by adding elements of $\mathcal{B}\text{.}$</p></li></ol><p id="p-110">If $U$ is a subspace of $V\text{,}$ then:</p><ol class="decimal"><li id="li-13"><p id="p-111">$U$ is finite-dimensional, and $\dim U\leq \dim V\text{.}$</p></li><li id="li-14"><p id="p-112">If $\dim U = \dim V\text{,}$ then $U=V\text{.}$</p></li></ol></article></div>

<div class="mathbook-content"><article class="exercise-like" id="exercise-15"><h6 class="heading"><span class="type">Exercise</span> <span class="codenumber">6.15</span>.</h6><p id="p-117">Give two examples of infinite-dimensional vector spaces. Support your answer.</p></article></div>

<div class="mathbook-content"><p id="p-118">Let's recap our results so far:</p><ul class="disc"><li id="li-15"><p id="p-119">A basis for a vector space $V$ is an independent set of vectors that spans $V\text{.}$</p></li><li id="li-16"><p id="p-120">The number of vectors in any basis of $V$ is a constant, called the dimension of $V\text{.}$</p></li><li id="li-17"><p id="p-121">The number of vectors in any independent set is always less than or equal to the number of vectors in a spanning set.</p></li><li id="li-18"><p id="p-122">In a finite-dimensional vector space, any independent set can be enlarged to a basis, and any spanning set can be cut down to a basis by deleting vectors that are in the span of the remaining vectors.</p></li></ul><p>Another important aspect of dimension is that it reduces many problems, such as determining equality of subspaces, to counting problems.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="thm-subspace-dim"><h6 class="heading"><span class="type">Theorem</span> <span class="codenumber">6.16</span>.</h6><p id="p-123">Let $U$ and $W$ be subspaces of a finite-dimensional vector space $V\text{.}$</p><ol class="decimal"><li id="li-19"><p id="p-124">If $U\subseteq W\text{,}$ then $\dim U\leq \dim W\text{.}$</p></li><li id="li-20"><p id="p-125">If $U\subseteq W$ and $\dim U=\dim W\text{,}$ then $U=W\text{.}$</p></li></ol></article><article class="proof" id="proof-3"><h6 class="heading"><span class="type">Proof.</span></h6><ol id="p-126" class="decimal"><li id="li-21"><p id="p-127">Suppose $U\subseteq W\text{,}$ and let $B=\{\vec{u}_1,\ldots, \vec{u}_k\}$ be a basis for $U\text{.}$ Since $B$ is a basis, it's independent. And since $B\subseteq U$ and $U\subseteq W\text{,}$ $B\subseteq W\text{.}$ Thus, $B$ is an independent subset of $W\text{,}$ and since any basis of $W$ spans $W\text{,}$ we know that $\dim U = k \leq \dim W\text{,}$ by <a href="sec-dimension.ipynb#theorem-steinitz" class="internal" title="Theorem 6.1: Fundamental Theorem (Steinitz Exchange Lemma)">Theorem 6.1</a>.</p></li><li id="li-22"><p id="p-128">Suppose $U\subseteq W$ and $\dim U = \dim W\text{.}$ Let $B$ be a basis for $U\text{.}$ As above, $B$ is an independent subset of $W\text{.}$ If $W\neq U\text{,}$ then there is some $\vec{w}\in W$ with $\vec{w}\notin U\text{.}$ But $U=\spn B\text{,}$ so that would mean that $B\cup \{\vec{w}\}$ is a linearly independent set containing $\dim U+1$ vectors. This is impossible, since $\dim W=\dim U\text{,}$ so no independent set can contain more than $\dim U$ vectors.</p></li></ol></article></div>

<div class="mathbook-content"><p id="p-129">An even more useful counting result is the following:</p></div>

<div class="mathbook-content"><article class="theorem-like" id="thm-half-the-work"><h6 class="heading"><span class="type">Theorem</span> <span class="codenumber">6.17</span>.</h6><p id="p-130">Let $V$ be an $n$-dimensional vector space. If the set $S$ contains $n$ vectors, then $S$ is independent if and only if $\spn S=V\text{.}$</p></article><article class="proof" id="proof-4"><h6 class="heading"><span class="type">Proof.</span></h6><p id="p-131">If $S$ is independent, then it can be extended to a basis $B$ with $S\subseteq B\text{.}$ But $S$ and $B$ both contain $n$ vectors (since $\dim V=n$), so we must have $S=B\text{.}$</p><p id="p-132">If $S$ spans $V\text{,}$ then $S$ must contain a basis $B\text{,}$ and as above, since $S$ and $B$ contain the same number of vectors, they must be equal.</p></article></div>

<div class="mathbook-content"><h6 class="heading"><span class="title">New subspaces from old.</span></h6></div>

<div class="mathbook-content"><p id="p-133">On your first assignment, you showed that if $U$ and $W$ are subspaces of a vector space $V\text{,}$ then the intersection $U\cap W$ is also a subspace of $V\text{.}$ You also showed that the union $U\cup W$ is generally not a subspace, unless one subspace is contained in the other (in which case the union is just the larger of the two subspaces we already have).</p></div>

<div class="mathbook-content"><p id="p-134">In class, we discussed the fact that the right way to define a subspace containing both $U$ and $W$ is using their <dfn class="terminology">sum</dfn>: we define the sum $U+W$ of two subspaces by</p><div class="displaymath">
\begin{equation*}
U+W = \{\vec{u}+\vec{w} \,|\, \vec{u}\in U \text{ and } \vec{w}\in W\}\text{.}
\end{equation*}
</div><p>We proved that $U+W$ is again a subspace of $V\text{.}$</p></div>

<div class="mathbook-content"><p id="p-135">If $U\cap W = \{\vec{0}\}\text{,}$ we say that the sum is a <dfn class="terminology">direct sum</dfn>, and write it as $U\oplus W\text{.}$ The following theorem might help us understand why direct sums are singled out for special attention:</p></div>

<div class="mathbook-content"><article class="theorem-like" id="thm-sum-dimension"><h6 class="heading"><span class="type">Theorem</span> <span class="codenumber">6.18</span>.</h6><p id="p-136">Let $U$ and $W$ be subspaces of a finite-dimensional vector space $V\text{.}$ Then $U+W$ is finite-dimensional, and</p><div class="displaymath">
\begin{equation*}
\dim(U+W)=\dim U + \dim W - \dim(U\cap W)\text{.}
\end{equation*}
</div></article></div>

<div class="mathbook-content"><p id="p-137">If the sum is direct, then we have simply $\dim(U\oplus W) = \dim U + \dim W\text{.}$ The other reason why direct sums are preferable, is that any $\vec{v}\in U\oplus W$ can be written <em class="emphasis">uniquely</em> as $\vec{v}=\vec{u}+\vec{w}$ where $\vec{U}\in U$ and $\vec{w}\in W\text{.}$</p></div>

<div class="mathbook-content"><article class="theorem-like" id="thm-direct-sum"><h6 class="heading"><span class="type">Theorem</span> <span class="codenumber">6.19</span>.</h6><p id="p-138">For any subspaces $U,W$ of a vector space $V\text{,}$ $U\cap W = \{\vec{0}\}$ if and only if for every $\vec{v}\in U+W$ there exist unique $\vec{u}\in U, \vec{w}\in W$ such that $\vec{v}=\vec{u}+\vec{w}\text{.}$</p></article><article class="proof" id="proof-5"><h6 class="heading"><span class="type">Proof.</span></h6><p id="p-139">Suppose that $U\cap W = \{\vec{0}\}\text{,}$ and suppose that we have $\vec{v} = \vec{u}_1+\vec{w}_1 = \vec{u}_2+\vec{w}_2\text{,}$ for $\vec{u}_1,\vec{u}_2\in U,\vec{w}_1,\vec{w}_2\in W\text{.}$ Then $\vec{0}=(\vec{u}_1-\vec{u}_2)+(\vec{w}_1-\vec{w}_2)\text{,}$ which implies that</p><div class="displaymath">
\begin{equation*}
\vec{w}_1-\vec{w}_2 = -(\vec{u}_1-\vec{u}_2)\text{.}
\end{equation*}
</div><p>Now, $\vec{u}=\vec{u}_1-\vec{u}_2\in U\text{,}$ since $U$ is a subspace, and similarly, $\vec{w}=\vec{w}_1-\vec{w}_2\in W\text{.}$ But we also have $\vec{w}=-\vec{u}\text{,}$ which implies that $\vec{w}\in U\text{.}$ Therefore, $\vec{w}\in U\cap W\text{,}$ which implies that $\vec{w}=\vec{0}\text{,}$ so $\vec{w}_1=\vec{w}_2\text{.}$ But we must also then have $\vec{u}=\vec{0}\text{,}$ so $\vec{u}_1=\vec{u}_2\text{.}$</p><p id="p-140">Conversely, suppose that every $\vec{v}\in U+W$ can be written uniquely as $\vec{v}=vec{u}+\vec{w}\text{,}$ with $\vec{u}\in U$ and $\vec{w}\in W\text{.}$ Suppose that $\vec{a}\in U\cap W\text{.}$ Then $\vec{a}\in U$ and $\vec{a}\in W\text{,}$ so we also have $-\vec{a}\in W\text{,}$ since $W$ is a subspace. But then $\vec{0}=\vec{a}+(-\vec{a})\text{,}$ where $\vec{a}\in U$ and $-\vec{a}\in W\text{.}$ On the other hand, $\vec{0}=\vec{0}+\vec{0}\text{,}$ and $\vec{0}$ belongs to both $U$ and $W\text{.}$ It follows that $\vec{a}=\vec{0}\text{.}$ Since $\vec{a}$ was arbitrary, $U\cap W = \{\vec{0}\}\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-141">We end with one last application of the theory we've developed on the existence of a basis for a finite-dimensional vector space. As we continue on to later topics, we'll find that it is often useful to be able to decompose a vector space into a direct sum of subspaces. Using bases, we can show that this is always possible.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="thm-construct-complement"><h6 class="heading"><span class="type">Theorem</span> <span class="codenumber">6.20</span>.</h6><p id="p-142">Let $V$ be a finite-dimensonal vector space, and let $U$ be any subspace of $V\text{.}$ Then there exists a subspace $W\subseteq V$ such that $U\oplus W = V\text{.}$</p><p id="p-146">The subspace $W$ constructed in the theorem above is called a <dfn class="terminology">complement</dfn> of $U\text{.}$ It is not unique; indeed, it depends on the choice of basis vectors. For example, if $U$ is a one-dimensional subspace of $\R^2\text{;}$ that is, a line, then any other non-parallel line through the origin provides a complement of $U\text{.}$ Later we will see that an especially useful choice of complement is the <dfn class="terminology">orthogonal complement</dfn>.</p></article><article class="proof" id="proof-6"><h6 class="heading"><span class="type">Proof.</span></h6><p id="p-143">Let $\{\vec{u}_1,\ldots, \vec{u}_m\}$ be a basis of $U\text{.}$ Since $U\subseteq W\text{,}$ the set $\{\vec{u}_1,\ldots, \vec{u}_m\}$ is a linearly independent subset of $V\text{.}$ Since any linearly independent set can be extended to a basis of $V\text{,}$ there exist vectors $\vec{w}_1,\ldots,\vec{w}_n$ such that</p><div class="displaymath">
\begin{equation*}
\{\vec{u}_1,\ldots, \vec{u}_m,\vec{w}_1,\ldots, \vec{w}_n\}
\end{equation*}
</div><p>is a basis of $V\text{.}$</p><p id="p-144">Now, let $W = \spn\{\vec{w}_1,\ldots, \vec{w}_n\}\text{.}$ Then $W$ is a subspace, and $\{\vec{w}_1,\ldots, \vec{w}_n\}$ is a basis for $W\text{.}$ (It spans, and must be independent since it's a subset of an independent set.)</p><p id="p-145">Clearly, $U+W=V\text{,}$ since $U+W$ contains the basis for $V$ we've constructed. To show the sum is direct, it suffices to show that $U\cap W = \{\vec{0}\}\text{.}$ To that end, suppose that $\vec{v}\in U\cap W\text{.}$ Since $\vec{v}\in U\text{,}$ we have</p><div class="displaymath">
\begin{equation*}
\vec{v}=a_1\vec{u}_1+\cdots +a_m\vec{u}_m
\end{equation*}
</div><p>for scalars $a_1,\ldots, a_m\text{.}$ Since $\vec{v}\in W\text{,}$ we can write</p><div class="displaymath">
\begin{equation*}
\vec{v}=b_1\vec{w}_1+\cdots + b_n\vec{w}_n
\end{equation*}
</div><p>for scalars $b_1,\ldots, b_n\text{.}$ But then</p><div class="displaymath">
\begin{equation*}
\vec{0}=\vec{v}-\vec{v}=a_1\vec{u}_1+\cdots a_m\vec{u}_m-b_1\vec{w}_1-\cdots -b_n\vec{w}_n.
\end{equation*}
</div><p>Since $\{\vec{u}_1,\ldots, \vec{u}_m,\vec{w}_1,\ldots, \vec{w}_n\}$ is a basis for $V\text{,}$ it's independent, and therefore, all of the $a_i,b_j$ must be zero, and therefore, $\vec{v}=\vec{0}\text{.}$</p></article></div>
