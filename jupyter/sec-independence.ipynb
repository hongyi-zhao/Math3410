<div class="mathbook-content"><h6 class="heading hide-type"><span class="type">Section</span> <span class="codenumber">2.3</span> <span class="title">Linear Independence</span></h6><a href="sec-independence.ipynb" class="permalink">Â¶</a></div>

<div class="mathbook-content"><p id="p-58">In any vector space $V\text{,}$ we say that a set of vectors</p><div class="displaymath">
\begin{equation*}
\{\vec{v}_1,\ldots,\vec{v}_2\}
\end{equation*}
</div><p>is <dfn class="terminology">linearly independent</dfn> if for any scalars $c_1,\ldots, c_k$</p><div class="displaymath">
\begin{equation*}
c_1\vec{v}_1+\cdots + c_k\vec{v}_k = \vec{0} \quad\Rightarrow\quad c_1=\cdots = c_k=0\text{.}
\end{equation*}
</div></div>

<div class="mathbook-content"><p id="p-59">This means that no vector in the set can be written as a linear combination of the other vectors in that set. We will soon see that when looking for vectors that span a subspace, it is especially useful to find a spanning set that is also linearly independent. The following lemma establishes some basic properties of independent sets.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-1"><h6 class="heading"><span class="type">Lemma</span> <span class="codenumber">2.3.1</span>.</h6><p id="p-60">In any vector space $V\text{:}$</p><ol class="decimal"><li id="li-1"><p id="p-61">If $\vec{v}\neq\vec{0}\text{,}$ then $\{\vec{v}\}$ is indenpendent.</p></li><li id="li-2"><p id="p-62">If $S\subseteq V$ contains the zero vector, then $S$ is dependent.</p></li></ol></article></div>

<div class="mathbook-content"><p id="p-63">The definition of linear independence tells us that if $\{\vec{v}_1,\ldots, \vec{v}_k\}$ is an independent set of vectors, then there is only one way to write $\vec{0}$ as a linear combination of these vectors; namely,</p><div class="displaymath">
\begin{equation*}
\vec{0} = 0\vec{v}_1+0\vec{v}_2+\cdots +0\vec{v_k}\text{.}
\end{equation*}
</div><p>In fact, more is true: <em class="emphasis">any</em> vector in the span of a linearly independent set can be written in only one way as a linear combination of those vectors.</p></div>

<div class="mathbook-content"><p id="p-64">Computationally, questions about linear independence are just questions about homogeneous systems of linear equations. For example, suppose we want to know if the vectors</p><div class="displaymath">
\begin{equation*}
\vec{u}=\bbm 1\\-1\\4\ebm, \vec{v}=\bbm 0\\2\\-3\ebm, \vec{w}=\bbm 4\\0\\-3\ebm
\end{equation*}
</div><p>are linearly independent in $\mathbb{R}^3\text{.}$ This question leads to the vector equation</p><div class="displaymath">
\begin{equation*}
x\vec{u}+y\vec{v}+z\vec{w}=\vec{0}\text{,}
\end{equation*}
</div><p>which becomes the matrix equation</p><div class="displaymath">
\begin{equation*}
\bbm 1\amp0\amp4\\-1\amp2\amp0\\4\amp-3\amp-3\ebm\bbm x\\y\\z\ebm = \bbm 0\\0\\0\ebm\text{.}
\end{equation*}
</div></div>

<div class="mathbook-content"><p id="p-65">We now apply some basic theory from linear algebra. A unique (and therefore, trivial) solution to this system is guaranteed if the matrix $A = \bbm 1\amp0\amp4\\-1\amp2\amp0\\4\amp-3\amp-3\ebm$ is invertible, since in that case we have $\bbm x\\y\\z\ebm = A^{-1}\vec{0} = \vec{0}\text{.}$</p></div>

<div class="mathbook-content"><p id="p-66">This approach is problematic, however, since it won't work if we have 2 vectors, or 4. Instead, we look at the reduced row-echelon form. A unique solution corresponds to having a leading 1 in each column of $A\text{.}$ Let's check this condition.</p></div>

```python
from sympy import *
init_printing()
```

```python
A = Matrix(3,3,[1,0,4,-1,2,0,4,-3,-3])
A.rref()
```

<div class="mathbook-content"><p id="p-67">One observation is useful here, and will lead to a better understanding of independence. First, it would be impossible to have 4 or more linearly independent vectors in $\mathbb{R}^3\text{.}$ Why? (How many leading ones can you have in a $3\times 4$ matrix?) Second, having two or fewer vectors makes it more likely that the set is independent.</p></div>

<div class="mathbook-content"><p id="p-68">The largest set of linearly independent vectors possible in $\mathbb{R}^3$ contains three vectors. You might have also observed that the smallest number of vectors needed to span $\mathbb{R}^3$ is 3. Hmm. Seems like there's something interesting going on here. But first, some more computation.</p></div>

<div class="mathbook-content"><article class="exercise-like" id="exercise-5"><h6 class="heading"><span class="type">Exercise</span> <span class="codenumber">2.3.2</span>.</h6><p id="p-69">Determine whether the set $\left\{\bbm 1\\2\\0\ebm, \bbm -1\\0\\3\ebm,\bbm -1\\4\\9\ebm\right\}$ is linearly independent in $\R^3\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-70">Again, we set up a matrix and reduce:</p></div>

```python
A = Matrix(3,3,[1,-1,-1,2,0,4,0,3,9])
A.rref()
```

<div class="mathbook-content"><p id="p-71">Notice that this time we don't get a unique solution, so we can conclude that these vectors are <em class="emphasis">not</em> independent. Furthermore, you can probably deduce from the above that we have $2\vec{v}_1+3\vec{v}_2-\vec{v}_3=\vec{0}\text{.}$ Now suppose that $\vec{w}\in\spn\{\vec{v}_1,\vec{v}_2,\vec{v}_3\}\text{.}$ In how many ways can we write $\vec{w}$ as a linear combination of these vectors?</p></div>

<div class="mathbook-content"><article class="exercise-like" id="exercise-6"><h6 class="heading"><span class="type">Exercise</span> <span class="codenumber">2.3.3</span>.</h6><p id="p-72">Which of the following subsets of $P_2(\mathbb{R})$ are independent?</p><div class="displaymath">
\begin{gather*}
\text{(a) } S_1 = \{x^2+1, x+1, x\}\\
\text{(b) } S_2 = \{x^2-x+3, 2x^2+x+5, x^2+5x+1\}
\end{gather*}
</div></article></div>

<div class="mathbook-content"><p id="p-73">In each case, we set up the defining equation for independence, collect terms, and then analyze the resulting system of equations. (If you work with polynomials often enough, you can probably jump straight to the matrix. For now, let's work out the details.)</p></div>

<div class="mathbook-content"><p id="p-74">Suppose</p><div class="displaymath">
\begin{equation*}
r(x^2+1)+s(x+1)+tx = 0\text{.}
\end{equation*}
</div><p>Then $rx^2+(s+t)x+(r+s)=0=0x^2+0x+0\text{,}$ so</p><div class="displaymath">
\begin{align*}
r \amp =0\\
s+t \amp =0\\
r+s\amp =0\text{.}
\end{align*}
</div><p>And in this case, we don't even need to ask the computer. The first equation gives $r=0$ right away, and putting that into the third equation gives $s=0\text{,}$ and the second equation then gives $t=0\text{.}$</p></div>

<div class="mathbook-content"><p id="p-75">Since $r=s=t=0$ is the only solution, the set is independent.</p></div>

<div class="mathbook-content"><p id="p-76">Repeating for $S_2$ leads to the equation</p><div class="displaymath">
\begin{equation*}
(r+2s+t)x^2+(-r+s+5t)x+(3r+5s+t)1=0.
\end{equation*}
</div><p>This gives us:</p></div>

```python
A = Matrix(3,3,[1,2,1,-1,1,5,3,5,1])
A.rref()
```

\left\{\bbm -1\amp 0\\0\amp -1\ebm, \bbm 1\amp -1\\ -1\amp 1\ebm,

\end{equation*}
</div><p>is linearly independent in $M_2(\mathbb{R})\text{.}$</p></article></div>

<div class="mathbook-content"><p id="p-78">Again, we set a linear combination equal to the zero vector, and combine:</p><div class="displaymath">
\begin{align*}
a\bbm -1\amp 0\\0\amp -1\ebm +b\bbm 1\amp -1\\ -1\amp 1\ebm
+c\bbm 1\amp 1\\1\amp 1\ebm +d \bbm 0\amp -1\\-1\amp 0\ebm = \bbm 0\amp 0\\ 0\amp 0\ebm\\
\bbm -a+b+c\amp -b+c-d\\-b+c-d\amp -a+b+c\ebm = \bbm 0\amp 0\\0\amp 0\ebm\text{.}
\end{align*}
</div></div>

<div class="mathbook-content"><p id="p-79">We could proceed, but we might instead notice right away that equations 1 and 4 are identical, and so are equations 2 and 3. With only two distinct equations and 4 unknowns, we're certain to find nontrivial solutions.</p></div>
