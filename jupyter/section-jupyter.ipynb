Some cells contain text, which can be in either <abbr class="initialism">HTML</abbr> or <dfn class="terminology">Markdown</dfn>. Markdown is a simple markup language. Use single dollar signs for inline math: <code class="code-inline tex2jax_ignore">$\frac{d}{dx}\sin(x)=\cos(x)$</code> produces $\frac{d}{dx}\sin(x)=\cos(x)\text{,}$ for example.</p></div>

\begin{equation*}
A = \begin{bmatrix}1\amp 2\amp 3\\4\amp 5\amp 6\end{bmatrix}\text{.}
\end{equation*}
