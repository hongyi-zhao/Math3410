<div class="mathbook-content"><h6 class="heading hide-type"><span xmlns:svg="http://www.w3.org/2000/svg" class="type">Section</span> <span class="codenumber">1.1</span> <span class="title">Abstract vector spaces</span></h6></div>

<div class="mathbook-content"><article class="definition definition-like" id="def-vector-space"><h6 xmlns:svg="http://www.w3.org/2000/svg" class="heading"><span class="type">Definition</span><span class="space"> </span><span class="codenumber">1.1.1</span><span class="period">.</span></h6><p id="p-11">A <dfn class="terminology">real vector space</dfn> (or vector space over $\R$) is a nonempty set $V\text{,}$ whose objects are called <dfn class="terminology">vectors</dfn>, equipped with two operations:</p><ol class="decimal"><li id="li-1"><p id="p-12"><dfn class="terminology">Addition</dfn>, which is a map from $V\times V$ to $V$ that associates each ordered pair of vectors $(\vv,\ww)$ to a vector $\vv+\ww\text{,}$ called the <dfn class="terminology">sum</dfn> of $\vv$ and $\ww\text{.}$</p></li><li id="li-2"><p id="p-13"><dfn class="terminology">Scalar multiplication</dfn>, which is a map from $\R\times V$ to $V$ that associates each real number $c$ and vector $\vv$ to a vector $c\vv\text{.}$</p></li></ol><p id="p-14">The operations of addition and scalar multiplication are required to satisfy the following <em class="emphasis">axioms</em>:</p><dl class="description-list"><dt id="li-3">A1.</dt><dd><p id="p-15">If $\uu,\vv\in V\text{,}$ then $\uu+\vv\in V\text{.}$ (Closure under addition)</p></dd><dt id="li-4">A2.</dt><dd><p id="p-16">For all $\uu,\vv\in V\text{,}$ $\uu+\vv=\vv+\uu\text{.}$ (Commutativity of addition)</p></dd><dt id="li-5">A3.</dt><dd><p id="p-17">For all $\uu,\vv,\ww\in V\text{,}$ $\uu+(\vv+\ww)=(\uu+\vv)+\ww\text{.}$ (Associativity of addition)</p></dd><dt id="li-6">A4.</dt><dd><p id="p-18">There exists an element $\zer\in V$ such that $\vv+\zer=\vv$ for each $\vv\in V\text{.}$ (Existence of a zero vector)</p></dd><dt id="li-7">A5.</dt><dd><p id="p-19">For each $\vv\in V\text{,}$ there exists a vector $-\vv\in V$ such that $\vv+(-\vv)=\zer\text{.}$ (Existence of negatives)</p></dd><dt id="li-8">S1.</dt><dd><p id="p-20">If $\vv\in V\text{,}$ then $c\vv\in V$ for all $c\in\R\text{.}$ (Closure under scalar multiplication)</p></dd><dt id="li-9">S2.</dt><dd><p id="p-21">For all $c\in \R$ and $\vv,\ww\in V\text{,}$ $c(\vv+\ww)=c\vv+c\ww\text{.}$ (Distribution over vector addition)</p></dd><dt id="li-10">S3.</dt><dd><p id="p-22">For all $a,b\in\R$ and $\vv\in V\text{,}$ $(a+b)\vv=a\vv+b\vv\text{.}$ (Distribution over scalar addition)</p></dd><dt id="li-11">S4.</dt><dd><p id="p-23">For all $a,b\in \R$ and $\vv\in V\text{,}$ $a(b\vv)=(ab)\vv\text{.}$ (Associativity of scalar multiplication)</p></dd><dt id="li-12">S5.</dt><dd><p id="p-24">For all $\vv\in V\text{,}$ $1\vv=\vv\text{.}$ (Normalization of scalar multiplication)</p></dd></dl></article></div>

<div class="mathbook-content"><p xmlns:svg="http://www.w3.org/2000/svg" id="p-25">Note that a zero vector must exist in every vector space. This simple observation is a key component of many proofs and counterexamples in linear algebra. In general, we may define a vector space whose scalars belong to a <em class="emphasis">field</em> $\mathbb{F}\text{.}$ A field is a set of objects whose algebraic properties are modelled after those of the real numbers. Fields have their <a class="external" href="https://en.wikipedia.org/wiki/Field%5F(mathematics)#Classic%5Fdefinition" target="_blank">own set of axioms</a>, which we will not list here. While it is possible to study linear algebra over <em class="emphasis">finite fields</em> (like the integers modulo a prime number) we will only consider two fields: the real numbers $\R\text{,}$ and the complex numbers $\C\text{.}$</p></div>

<div class="mathbook-content"><p xmlns:svg="http://www.w3.org/2000/svg" id="p-26">A vector space whose scalars are complex numbers will be called a <em class="emphasis">complex vector space</em>. While many students are initially intimidated by the complex numbers, most results in linear algebra work exactly the same over $\C$ as they do over $\R\text{.}$ And where the results differ, things are usually <em class="emphasis">easier</em> with complex numbers, owing in part to the fact that all complex polynomials can be completely factored.</p></div>

<div class="mathbook-content"><p xmlns:svg="http://www.w3.org/2000/svg" id="p-27">To help us gain familiarity with the abstract nature of <a href="sec-vec-sp.ipynb#def-vector-space" class="internal" title="Definition 1.1.1">DefinitionÂ 1.1.1</a>, let us consider some basic examples.</p></div>

<div class="mathbook-content"><article class="example example-like" id="ex-vector-spaces"><h6 xmlns:svg="http://www.w3.org/2000/svg" class="heading"><span class="type">Example</span><span class="space"> </span><span class="codenumber">1.1.2</span><span class="period">.</span></h6><p id="p-28">The following are examples of vector spaces. We leave verification of axioms as an exercise.</p><ol class="decimal"><li id="li-13"><p id="p-29">The set $\R^n$ of $n$-tuples $(x_1,x_2,\ldots, x_n)$ of real numbers, where we define</p><div class="displaymath">
\begin{align*}
(x_1,x_2,\ldots, x_n)+(y_1,y_2,\ldots, y_n) \amp = (x_1+y_1,x_2+y_2,\ldots, x_n+y_n) \\
c(x_1,x_2,\ldots, x_n)\amp = (cx_1,cx_2,\ldots, cx_n)\text{.}
\end{align*}
</div><p data-braille="continuation">We will also often use $\R^n$ to refer to the vector space of $1\times n$ column matrices $\bbm x_1\\x_2\\\vdots\\x_n\ebm\text{,}$ where addition and scalar multiplication are defined as for matrices (and the same as the above, with the only difference being the way in which we choose to write our vectors). If the distinction between $n$-tuples and column matrices is ever important, it will be made clear.</p></li><li id="li-14"><p id="p-30">The set $\mathbf{M}_{mn}$ of $m\times n$ matrices, equipped with the usual matrix addition and scalar multiplication.</p></li><li id="li-15"><p id="p-31">The set $\mathbf{P}_n$ of all polynomials</p><div class="displaymath">
\begin{equation*}
p(x) = a_0+a_1x+\cdots + a_nx^n
\end{equation*}
</div><p data-braille="continuation">of degree less than or equal to $n\text{,}$ where, for</p><div class="displaymath">
\begin{align*}
p(x) \amp = a_0+a_1x+\cdots + a_nx^n \\
q(x) \amp = b_0+b_1x+\cdots +b_nx^n
\end{align*}
</div><p data-braille="continuation">we define</p><div class="displaymath">
\begin{equation*}
p(x)+q(x)=(a_0+b_0)+(a_1+b_1)x+\cdots + (a_n+b_n)x^n
\end{equation*}
</div><p data-braille="continuation">and</p><div class="displaymath">
\begin{equation*}
cp(x) = ca_0+(ca_1)x+\cdots + (ca_n)x^n\text{.}
\end{equation*}
</div><p data-braille="continuation">The zero vector is the polynomial $0=0+0x+\cdots + 0x^n\text{.}$</p></li><li id="li-16"><p id="p-32">The set $\mathbf{F}[a,b]$ of all functions $f:[a,b]\to \R\text{,}$ where we define $(f+g)(x)=f(x)+g(x)$ and $(cf)(x)=c(f(x))\text{.}$ The zero function is the function satisfying $0(x)=0$ for all $x\in [a,b]\text{,}$ and the negative of a function $f$ is given by $(-f)(x)=-f(x)$ for all $x\in [a,b]\text{.}$</p></li></ol></article></div>

<div class="mathbook-content"><p xmlns:svg="http://www.w3.org/2000/svg" id="p-33">Other common examples of vector spaces can be found online; for example, <a class="external" href="https://en.wikipedia.org/wiki/Examples%5Fof%5Fvector%5Fspaces" target="_blank">on Wikipedia</a>. It is also interesting to try to think of less common examples. For example, can you think of a way to define a vector space structure on the set $(0,\infty)$ of all positive real numbers?</p></div>

<div class="mathbook-content"><p xmlns:svg="http://www.w3.org/2000/svg" id="p-34">There are a number of other algebraic properties that are common to all vector spaces; for example, it is true that $0\vv = \zer$ for all vectors $\vv$ in any vector space $V\text{.}$ The reason these are not included is that the ten axioms in <a href="sec-vec-sp.ipynb#def-vector-space" class="internal" title="Definition 1.1.1">DefinitionÂ 1.1.1</a> are the ones deemed âessentialâ â all other properties can be deduced from the axioms. To demonstrate, we next give the proof that $0\vv = \zer\text{.}$</p></div>

<div class="mathbook-content"><article class="theorem theorem-like" id="thm-zero-mult"><h6 xmlns:svg="http://www.w3.org/2000/svg" class="heading"><span class="type">Theorem</span><span class="space"> </span><span class="codenumber">1.1.3</span><span class="period">.</span></h6><p id="p-35">In any vector space $V\text{,}$ we have $0\vv = \zer$ for all $\vv\in V$</p></article><article class="proof" id="proof-1"><h6 class="heading"><span class="type">Proof<span class="period">.</span></span></h6><p id="p-36">Since $0+0=0\text{,}$ we have $0\vv=(0+0)\vv\text{.}$ Using the distributive axiom S3, this becomes</p><div class="displaymath">
\begin{equation*}
0\vv + 0\vv = 0\vv\text{.}
\end{equation*}
</div><p data-braille="continuation">By axiom A5, there is an element $-0\vv\in V$ such that $0\vv+(-0\vv)=\zer\text{.}$ Adding this to both sides of the equation above, we get:</p><div class="displaymath">
\begin{equation*}
(0\vv+0\vv)+(-0\vv) = 0\vv+(-0\vv)\text{.}
\end{equation*}
</div><p data-braille="continuation">Now, apply the associative property (A3) on the left, and A5 on the right, to get</p><div class="displaymath">
\begin{equation*}
0\vv + (0\vv+(-0\vv)) = \zer\text{.}
\end{equation*}
</div><p data-braille="continuation">Using A5 again on the left, we get $0\vv+\zer = \zer\text{.}$ Finally, axiom A4 guarantees $0\vv = 0\vv+\zer = \zer\text{.}$</p></article></div>

<div class="mathbook-content"><p xmlns:svg="http://www.w3.org/2000/svg" id="p-37">Similar tactics can be used to establish the following results, which we leave as an exercise. Solutions are included, but it will be worth your while in the long run to wrestle with these.</p></div>

<div class="mathbook-content"><article class="exercise exercise-like" id="ex-more-props"><h6 xmlns:svg="http://www.w3.org/2000/svg" class="heading"><span class="type">Exercise</span><span class="space"> </span><span class="codenumber">1.1.4</span><span class="period">.</span></h6><p id="p-38">Show that the following properties are valid in any vector space $V\text{:}$</p><ol class="decimal"><li id="li-17"><p id="p-39">If $\uu+\vv=\uu+\ww\text{,}$ then $\vv=\ww\text{.}$</p></li><li id="li-18"><p id="p-40">For any scalar $c\text{,}$ $c\zer=\zer\text{.}$</p></li><li id="li-19"><p id="p-41">If $c\vv=\zer\text{,}$ then either $c=0$ or $\vv=\zer\text{.}$</p></li><li id="li-20"><p id="p-42">For any $\vv\in V\text{,}$ $(-1)\vv=-\vv\text{.}$</p></li><li id="li-21"><p id="p-43">The zero vector is the unique vector such that $\vv+\zer=\vv$ for all $\vv\in V\text{.}$</p></li><li id="li-22"><p id="p-44">The negative $-\vv$ of any vector $\vv$ is unique.</p></li></ol><div class="solutions"><a data-knowl="" class="id-ref solution-knowl original" data-refid="hk-solution-1" id="solution-1"><span class="type">Solution</span></a><div class="hidden-content tex2jax_ignore" id="hk-solution-1"><div class="solution solution-like"><ol id="p-45" class="decimal"><li id="li-23"><p id="p-46">Suppose $\uu+\vv=\uu+\ww\text{.}$ By adding $-\uu$ on the left of each side, we obtain:</p><div class="displaymath">
\begin{align*}
-\uu+(\uu+\vv) \amp =-\uu+(\uu+\ww)\\
(-\uu+\uu)+\vv \amp =(-\uu+\uu)+\ww \quad \text{ by A3}\\
\zer+\vv \amp =\zer+\ww \quad \text{ by A5}\\
\vv \amp =\ww \quad \text{ by A4}\text{.}
\end{align*}
</div></li><li id="li-24"><p id="p-47">We have $c\zer = c(\zer+\zer) = c\zer +c\zer\text{,}$ by A4 and S2, respectively. Adding $-c\zer$ to both sides (and using axioms A3, A4, A5 as in the proof of <a href="sec-vec-sp.ipynb#thm-zero-mult" class="internal" title="Theorem 1.1.3">TheoremÂ 1.1.3</a>) we get $\zer = c\zer\text{.}$</p></li><li id="li-25"><p id="p-48">The main difficulty with this problem is getting the logic of the statement correct, and making sure you know what it is you're trying to prove. The desired conclusion is an <em class="emphasis">or</em> statement, which means it suffices to establish one part or the other. Typically, the way to proceed in these cases is to argue that if the first part is false, then the second must be true. This is how we proceed here.</p><p id="p-49">Suppose that $c\vv=\zer\text{.}$ If $c=0\text{,}$ then it's true that $c=0$ or $\vv=\zer\text{,}$ so there's nothing to prove. Suppose then that $c\neq 0\text{.}$ Then there exists a real number $\frac1c$ such that $c\bigl(\frac1c\bigr)=1\text{.}$ Multiplying both sides of $c\vv=\zer$ by $\frac1c\text{,}$ we get</p><div class="displaymath">
\begin{equation*}
\frac1c(c\vv)=\frac1c\zer\text{.}
\end{equation*}
</div><p data-braille="continuation">By the previous problem, we know that $\frac1c\zer = \zer\text{,}$ and by axioms S4 and S5, we have</p><div class="displaymath">
\begin{equation*}
\frac1c(c\vv)=\bigl(\frac1c\cdot c\bigr)\vv = 1\vv = \vv\text{,}
\end{equation*}
</div><p data-braille="continuation">showing that $\vv=\zer\text{,}$ and thus $c=0$ or $\vv=\zer\text{.}$ (This is a proof by cases, relying on the tautology $c=0$ or $c\neq 0\text{.}$)</p></li><li id="li-26"><p id="p-50">Suppose there are two vectors $\zer_1,\zer_2$ that act as additive identities. Then</p><div class="displaymath">
\begin{align*}
\zer_1 \amp = \zer_1+\zer_2 \quad \text{ since } \vv+\zer_2=\vv \text{ for any } \vv\\
\amp =\zer_2+\zer_1 \quad \text{ by axiom A2}\\
\amp \zer_2 \quad \text{ since } \vv+\zer_1=\vv \text{ for any } \vv
\end{align*}
</div><p data-braille="continuation">So any two vectors satisfying the property in A4 must, in fact, be the same.</p></li><li id="li-27"><p id="p-51">Let $\vv\in V\text{,}$ and suppose there are vectors $\ww_1,\ww_2\in V$ such that $\vv+\ww_1=\zer$ and $\vv+\ww_2=\zer\text{.}$ Then</p><div class="displaymath">
\begin{align*}
\ww_1 \amp = \ww_1+\zer \quad  \text{ by A4}\\
\amp = \ww_1+(\vv+\ww_2) \quad \text{ by assumption}\\
\amp = (\ww_1+\vv)+\ww_2 \quad \text{ by A3}\\
\amp = (\vv+\ww_1)+\ww_2 \quad \text{ by A2}\\
\amp = \zer+\ww_2 \quad \text{ by assumption}\\
\amp \ww_2 \quad \text{ by A4}\text{.}
\end{align*}
</div></li></ol></div></div></div></article></div>
